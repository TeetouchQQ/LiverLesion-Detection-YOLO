```python
def pascal_voc_to_yolo(x1, y1, x2, y2, image_w, image_h):
    return ((x2 + x1)/(2*image_w)), ((y2 + y1)/(2*image_h)), (x2 - x1)/image_w, (y2 - y1)/image_h
```

```python

from bs4 import BeautifulSoup
import re
 
# Reading the data inside the xml
# file to a variable under the name
# data
with open('/content/yolov7/runs/detect/exp3/labels/00001.xml', 'r') as f:
    data = f.read()
 
# Passing the stored data inside
# the beautifulsoup parser, storing
# the returned object
Bs_data = BeautifulSoup(data, "xml")
 
# Finding all instances of tag
# `unique`
b_unique = Bs_data.find_all('bndbox')
b_name = Bs_data.find_all('name')
print(len(b_unique))
#print(b_name)
for i in range(len(b_unique)):
  name = ''.join(re.findall(r'[A-Z]', str(b_name[i])))
  print(name)
  xmin = str(b_unique[i].find('xmin'))
  ymin = str(b_unique[i].find('ymin'))
  xmax = str(b_unique[i].find('xmax'))
  ymax = str(b_unique[i].find('ymax'))
  print(re.findall(r'\d+',xmin)[0])
  print(re.findall(r'\d+',ymin)[0])
  print(re.findall(r'\d+',xmax)[0])
  print(re.findall(r'\d+',ymax)[0])
  print('=================')
```

```python
###LOOP VOC
answer_dict = {}
for idx in range(2500):
  
  nname = str(idx).zfill(5) + '.jpg'
  idx2 = str(idx)
  idx2 = idx2.zfill(5)
  file_path = '/content/yolov7/runs/detect/exp4/labels/' + idx2 + '.xml'
  print(file_path)
  try:
    with open(file_path, 'r') as f:
        data = f.read()
    
    # Passing the stored data inside
    # the beautifulsoup parser, storing
    # the returned object
    Bs_data = BeautifulSoup(data, "xml")
    
    # Finding all instances of tag
    # `unique`
    b_unique = Bs_data.find_all('bndbox')
    b_name = Bs_data.find_all('name')
    print(len(b_unique))
    #print(b_name)
    objectB = []
    for i in range(len(b_unique)):
      name = str(b_name[i]).replace('name','')
      name = name.replace('<>','')
      name = name.replace('</>','')
      #name = name.replace('/<name>','')

      #ymin xmin ymax xmax
      print(name)
      xmin = str(b_unique[i].find('xmin'))
      ymin = str(b_unique[i].find('ymin'))
      xmax = str(b_unique[i].find('xmax'))
      ymax = str(b_unique[i].find('ymax'))

      xmin = int(re.findall(r'\d+',xmin)[0])
      ymin = int(re.findall(r'\d+',ymin)[0])
      xmax= int(re.findall(r'\d+',xmax)[0])
      ymax = int(re.findall(r'\d+',ymax)[0])
      print('=================')
      objectB.append([ymin,xmin,ymax,xmax,name])
    answer_dict[nname] = objectB
  except:
    answer_dict[nname] = []
```
