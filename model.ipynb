```python
from pylabel import importer
path_to_annotations = "/content/yolov7/runs/detect/exp4/labels"

#Identify the path to get from the annotations to the images 
path_to_images = "/content/yolov7/runs/detect/exp4"

#Import the dataset into the pylable schema 
#Class names are defined here https://github.com/ultralytics/yolov5/blob/master/data/coco128.yaml
yoloclasses = ['cystic','FFS','solid']
dataset = importer.ImportYoloV5(path=path_to_annotations, path_to_images=path_to_images, cat_names=yoloclasses,
    img_ext="jpg", name="test")

dataset.df.head(5)

print(f"Number of images: {dataset.analyze.num_images}")
print(f"Number of classes: {dataset.analyze.num_classes}")
print(f"Classes:{dataset.analyze.classes}")
print(f"Class counts:\n{dataset.analyze.class_counts}")
```

```python
###LOOP VOC
from bs4 import BeautifulSoup
import re
import json
answer_dict = {}
for idx in range(2500):
  
  nname = str(idx).zfill(5) + '.jpg'
  idx2 = str(idx)
  idx2 = idx2.zfill(5)
  file_path = '/content/yolov7/runs/detect/exp4/labels/' + idx2 + '.xml'
  print(file_path)
  try:
    with open(file_path, 'r') as f:
        data = f.read()
    
    # Passing the stored data inside
    # the beautifulsoup parser, storing
    # the returned object
    Bs_data = BeautifulSoup(data, "xml")
    
    # Finding all instances of tag
    # `unique`
    b_unique = Bs_data.find_all('bndbox')
    b_name = Bs_data.find_all('name')
    print(len(b_unique))
    #print(b_name)
    objectB = []
    for i in range(len(b_unique)):
      name = str(b_name[i]).replace('name','')
      name = name.replace('<>','')
      name = name.replace('</>','')
      #name = name.replace('/<name>','')

      #ymin xmin ymax xmax
      print(name)
      xmin = str(b_unique[i].find('xmin'))
      ymin = str(b_unique[i].find('ymin'))
      xmax = str(b_unique[i].find('xmax'))
      ymax = str(b_unique[i].find('ymax'))

      xmin = int(re.findall(r'\d+',xmin)[0])
      ymin = int(re.findall(r'\d+',ymin)[0])
      xmax= int(re.findall(r'\d+',xmax)[0])
      ymax = int(re.findall(r'\d+',ymax)[0])
      print('=================')
      objectB.append([xmin,ymin,xmax,ymax,name])
    answer_dict[nname] = objectB
  except:
    answer_dict[nname] = []
```
